```python
# 1. ÌïÑÏàò Ìå®ÌÇ§ÏßÄ ÏÑ§Ïπò
!pip install ultralytics yt-dlp opencv-python --quiet

# 2. Ïú†ÌäúÎ∏å ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú
import yt_dlp
import glob

video_url = "https://www.youtube.com/shorts/0N0iZudj0R4"
ydl_opts = {
    'format': 'bestvideo+bestaudio/best',
    'outtmpl': 'input_video.%(ext)s',
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

# 3. Îã§Ïö¥Î°úÎìúÎêú ÌååÏùºÎ™Ö ÏûêÎèô Ïù∏Ïãù
video_files = glob.glob("input_video.*")
if video_files:
    input_video_path = video_files[0]
    print(f"üé• Îã§Ïö¥Î°úÎìú ÏôÑÎ£å: {input_video_path}")
else:
    raise FileNotFoundError("‚ùå Îã§Ïö¥Î°úÎìúÎêú ÎπÑÎîîÏò§ ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")

# 4. YOLOv8 Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
from ultralytics import YOLO
model = YOLO("yolov8n.pt")  # Í∞ÄÏû• ÏûëÏùÄ YOLOv8 Î™®Îç∏

# 5. ÎπÑÎîîÏò§ ÏùΩÍ≥† Ï≤òÎ¶¨
import cv2

cap = cv2.VideoCapture(input_video_path)
if not cap.isOpened():
    raise IOError(f"‚ùå ÎπÑÎîîÏò§ ÌååÏùº Ïó¥Í∏∞Ïóê Ïã§Ìå®ÌñàÏäµÎãàÎã§: {input_video_path}")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"üîç Ìï¥ÏÉÅÎèÑ: {width}x{height}, FPS: {fps}")

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("output_yolov8.mp4", fourcc, fps, (width, height))

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(source=frame, save=False, imgsz=640)
    annotated_frame = results[0].plot()
    out.write(annotated_frame)
    frame_count += 1

cap.release()
out.release()
print(f"‚úÖ Í∞ùÏ≤¥ Ïù∏Ïãù ÏôÑÎ£å! Ï¥ù {frame_count} ÌîÑÎ†àÏûÑ Ï≤òÎ¶¨Îê®.")

# 6. ColabÏóêÏÑú ÏòÅÏÉÅ ÌëúÏãú
from IPython.display import Video
Video("output_yolov8.mp4", embed=True)

# 7. Colab ‚Üí Î°úÏª¨Î°ú Îã§Ïö¥Î°úÎìú
from google.colab import files
files.download("output_yolov8.mp4")
```
